# Visualises Symmetric Nuclear Matter (SNM) data 

```python
import numpy as np # arrays etcetera
import matplotlib.pyplot as plt # general plotting
import matplotlib as mpl # for figure quality (dpi)
import os # Dealing with paths
```

```python
# Figure quality
mpl.rcParams['figure.dpi']= 100 # in notebook resolution
mpl.rc("savefig", dpi=400) # higher resolution when saving

# Formatting options
fs = 15 # fontsize
```

```python
# Path to project files (You may need to change this)
currentDir = os.getcwd()
#dataDir = os.path.join(currentDir, '../../../Data')
dataDir = currentDir+'/../../Data'

# Stored model variations
# Baseline scenario \Lambda = 900 MeV
VARIATION = dataDir+"/DLW/QMC/Bag_Mass_Param_Used/1_Standard/"

# Bag model param.  -> /Bag_Mass_Param_Used/1_Standard/
#1_Standard
#2_L1-0
#3_L1-1
#4_L1-2
#5_L1-3
#6_L1-1_gsy1-3
#7_L1-3_gsy1-3
#8_L2-0_gsy1-9
#9_Increased_f_div_g
#10_Fock_ds
#11_Eff_Pro_Mass
#12_Eff_Pro_Mass_L1-1
#13_Eff_Pro_Mass_ds
#14_Dirac_Only
#15_Hartree
#16_R0-8
#17_App_S0_32-5
#18_App_S0_30
#19_S0_30
#20_Standard_BETA_N_ONLY
#21_MyQMC700 


# NJL model param. -> /NJL_Mass_Param_Used/1_NJL_Hartree/

#1_NJL_Hartree
#2_NJL_Standard
#3_NJL_L1-3
#4_NJL_Dirac_Only
#5_NJL_Fs1

# Select the data you wish to visualise 
DATA_PATH = VARIATION+"SNM/"

# Name of Model for plots (ENTER THE NAME YOU WANT HERE)
MODEL = r"A model"

# Path to other peoples data 
OTHERS_DATA= dataDir+"/OTHERS/"
HIC = "Pawel_Danielewicz_HIC_PvsDensity_SNM_PNM/"
```

## Additional Data

### dan_snm.dat

This file contains Danielewicz et al. heavy ion collision (HIC) constraints for SNM. This data was shared by Jirina Stone. This file contains 2 columns: 

Total baryonic density $\rho$ [fm$^{-3}$] and pressure $P$ [MeVfm$^{-3}$].

```python
dan_snm = np.loadtxt(OTHERS_DATA+HIC+"dan_snm.dat")
print(dan_snm.shape)
```

Danielewics et al HIC data only.

```python
fig1 = plt.figure(figsize=(6,4))
# alpha ~ opacity 0.0 - 1.0
plt.fill(dan_snm[:,0],dan_snm[:,1], alpha=0.2,label="Danielewicz et al.")

# Labels
plt.xlabel(r"Density $\rho$   [fm$^{-3}$]", fontsize=fs)
plt.ylabel(r"Pressure $P$   [MeVfm$^{-3}$]",fontsize=fs)

# Legend
plt.legend(frameon=False, loc="upper left",fontsize=fs)

# Save it
plt.savefig("test1.pdf")
```

## SymmetricMatLong Option Data Files

There are two other data files that are created. They are $\texttt{model\_parameters.dat}$ and $\texttt{snm\_saturation\_properties.dat}$. These two data files tabulate various quantities that are either chosen and used for the calculation or are calculated. All quantities are defined within these data files, so nothing more will be said here.

### snm_eos.dat

This file contains 4 columns: 

Total baryonic density $\rho$ [fm$^{-3}$], pressure $P$ [MeVfm$^{-3}$], total energy density $\epsilon$ [MeVfm$^{-3}$] and binding energy $\mathcal{E}$ [MeV].

```python
data = np.loadtxt(DATA_PATH+"snm_eos.dat")
print(data.shape)
```

The first plot is just a regular plot of pressure versus total baryonic density including all data with Danielewicz et al HIC constraints overlayed.

```python
fig2 = plt.figure(figsize=(6,4))
plt.plot(data[:,0], data[:,1], label=MODEL)

# Limits
plt.ylim(0,900)

# Labels
plt.xlabel(r"Density $\rho$   [fm$^{-3}$]", fontsize=fs)
plt.ylabel(r"Pressure $P$   [MeVfm$^{-3}$]",fontsize=fs)

# Add Danielewicz data (constraint from heavy ion collisions)
# alpha ~ opacity 0.0 - 1.0
plt.fill(dan_snm[:,0],dan_snm[:,1], alpha=0.2,label="Danielewicz et al.") 

# Legend
plt.legend(frameon=False, loc="best",fontsize=fs)

# Save it
plt.savefig("test2.pdf")
```

The same plot as above, but the y-axis has a log scale. This looks much better. Also the x-axis limits are different.

```python
fig3 = plt.figure(figsize=(6,4))
plt.plot(data[:,0], data[:,1], label=MODEL,linestyle='dashed',linewidth=3.0)

# Limits
plt.xlim(0,0.8)
plt.ylim(1,900)

# Log scale
ax = plt.gca()
ax.set_yscale("log")

# Labels
plt.xlabel(r"Density $\rho$   [fm$^{-3}$]", fontsize=fs)
plt.ylabel(r"Pressure $P$   [MeVfm$^{-3}$]",fontsize=fs)

# Add Danielewicz data (constraint from heavy ion collisions)
# alpha ~ opacity 0.0 - 1.0
plt.fill(dan_snm[:,0],dan_snm[:,1], alpha=0.2,label="Danielewicz et al.",color='plum')
plt.legend(frameon=False, loc="best", fontsize=fs)

# Save it
plt.savefig("test3.pdf")
```

Binding energy versus total baryonic density plot.

```python
fig4 = plt.figure(figsize=(6,4))

# Saturation density and binding energy 
rsat=0.16
beSat =-15.865

plt.plot(data[:,0],data[:,3], label=MODEL)
plt.plot(rsat,beSat,'ro')

# Lines to mark saturation
plt.axvline(x=rsat,color='r',linestyle='--')
plt.axhline(y=beSat,color='r',linestyle='--')

# Limits
plt.xlim(0,0.4)
plt.ylim(-20,20)

# Labels
plt.xlabel(r"Density $\rho$ [fm$^{-3}$]",fontsize=fs)
plt.ylabel(r"Binding Energy $\mathcal{E}$ [MeV]",fontsize=fs)

# Legend
plt.legend(frameon=False, loc="best", fontsize=fs)

# Save it
fig4.savefig("test4.pdf")
```

```python
fig5 = plt.figure(figsize=(6,4))
plt.plot(data[:,2],data[:,1], label=MODEL)

# Labels
plt.xlabel(r"Energy Density $\epsilon$ [MeVfm$^{-3}$]",fontsize=fs)
plt.ylabel(r"Pressure $P$ [MeVfm$^{-3}$]",fontsize=fs)

# Legend
plt.legend(frameon=False, loc="best", fontsize=fs)

# Save it
fig5.savefig("test5.pdf")
```

### snm_symmetry_energy.dat and snm_slope.dat

These files both contain 2 columns:

- Total baryonic density $\rho$ [fm$^{-3}$] and symmetry energy $S(\rho)$ [MeV] (definition depends on option chosen either numerical derivative or difference formula).

- Total baryonic density $\rho$ [fm$^{-3}$] and slope $L(\rho) = 3\rho\frac{\partial S}{\partial \rho}$ [MeV].

```python
Symdata = np.loadtxt(DATA_PATH+"snm_symmetry_energy.dat")
print(Symdata.shape)
```

```python
Slopedata = np.loadtxt(DATA_PATH+"snm_slope.dat")
print(Slopedata.shape)
```

```python
# First plot
#============
fig6 = plt.figure(figsize=(6,9))
ax6A = fig6.add_subplot(211)

ax6A.plot(Symdata[:,0], Symdata[:,1], label=MODEL,linestyle='dashed',
          linewidth=3.0)

# Labels
ax6A.set_ylabel(r"Symmetry Energy $S(\rho)$   [MeV]",fontsize=fs)

# Make these tick labels invisible
plt.setp(ax6A.get_xticklabels(), visible=False)

# Legend
ax6A.legend(frameon=False, loc="best", fontsize=fs)

# Second plot
#=============
ax6B = fig6.add_subplot(212,sharex=ax6A)

ax6B.plot(Slopedata[:,0], Slopedata[:,1], label=MODEL,linestyle='dashed',
          linewidth=3.0)

# Labels
ax6B.set_xlabel(r"Density $\rho$   [fm$^{-3}$]", fontsize=fs)
ax6B.set_ylabel(r"Slope $L(\rho)$   [MeV]",fontsize=fs)

# Legend
ax6B.legend(frameon=False, loc="upper left", fontsize=fs)

# Figure adjustments
fig6.subplots_adjust( hspace=0.09)
fig6.subplots_adjust(top=0.94)

# Save it
fig6.savefig("test6.pdf")
```

## snm_Fock_energy_densities.dat

This file contains 5 columns:

Total baryonic density $\rho$ [fm$^{-3}$] and the Fock contributions to the energy density $\epsilon_{i}$ [MeVfm$^{-3}$] for each of the mesons $i\in \lbrace\sigma,\omega,\rho,\pi\rbrace$.

```python
eFockdata = np.loadtxt(DATA_PATH+"snm_Fock_energy_densities.dat")
print(eFockdata.shape)
```

```python
fig7 = plt.figure(figsize=(6,4))
plt.plot(eFockdata[:,0],eFockdata[:,1], label=r"$\epsilon_{\sigma}$")
plt.plot(eFockdata[:,0],eFockdata[:,2], label=r"$\epsilon_{\omega}$")
plt.plot(eFockdata[:,0],eFockdata[:,3], label=r"$\epsilon_{\rho}$")
plt.plot(eFockdata[:,0],eFockdata[:,4], label=r"$\epsilon_{\pi}$")

# Legend
plt.legend(frameon=False, loc="best", fontsize=fs)

# Labels
plt.xlabel(r"Density $\rho$ [fm$^{-3}$]",fontsize=fs)
plt.ylabel(r"Energy Density $\epsilon_{i}$ [MeVfm$^{-3}$]",fontsize=fs )

# Save it
fig7.savefig("test7.pdf")
```

### snm_N_Masses.dat

This file contains 6 columns:

Total baryonic density $\rho$ [fm$^{-3}$], effective masses $M^{\ast}_{i}$ [MeV], the difference $M^{\ast}_{n}-M^{\ast}_{p}$ [MeV], and the ratios $M^{\ast}_{i}/M_{i}$ where $i\in\lbrace n,p\rbrace$.

```python
Massesdata = np.loadtxt(DATA_PATH+"snm_N_Masses.dat")
print(Massesdata.shape)
```

```python
fig8 = plt.figure(figsize=(6,4))

# Density and effective nucleon mass at saturation
rsat=0.16
MStarSat = 7.855269E+02 #(ENTER VALUE HERE, LOOK IN snm_saturation_properties.dat FOR THE VARIATION)

# Lines to mark saturation
plt.axvline(x=rsat,color='g',linestyle='--')
plt.axhline(y=MStarSat,color='g',linestyle='--')
plt.plot(rsat,MStarSat,'go')

plt.plot(Massesdata[:,0],Massesdata[:,1],'r',label=r"$M_{n}^{\ast}$")
plt.plot(Massesdata[:,0],Massesdata[:,2],'b--',label=r"$M_{p}^{\ast}$")

# Legend
plt.legend(frameon=False, loc="best", fontsize=fs)

# Labels
plt.xlabel(r"Density $\rho$ [fm$^{-3}$]",fontsize=fs)
plt.ylabel(r"Nucleon Masses $M_{i}^{\ast}$ [MeV]",fontsize=fs )

# Save it
fig8.savefig("test8.pdf")
```

### snm_N_den_dep_coup.dat

This file contains 3 columns:
  
Total baryonic density $\rho$ [fm$^{-3}$] and the density dependent couplings $g_{\sigma N}C_{i}(\bar{\sigma}) = -\frac{\partial M^{\ast}_{i}}{\partial\bar{\sigma}}$, where $i\in\lbrace n, p\rbrace$.

```python
DenDepCCdata = np.loadtxt(DATA_PATH+"snm_N_den_dep_coup.dat")
print(DenDepCCdata.shape)
```

```python
fig9 = plt.figure(figsize=(6,4))
plt.plot(DenDepCCdata[:,0],DenDepCCdata[:,1],'r',label=r"$g_{\sigma N}C_{n}(\bar{\sigma})$")
plt.plot(DenDepCCdata[:,0],DenDepCCdata[:,2],'b--',label=r"$g_{\sigma N}C_{p}(\bar{\sigma})$")

# Labels
plt.xlabel(r"Density $\rho$ [fm$^{-3}$]",fontsize=fs)
plt.ylabel(r"Density Dependent Coupling",fontsize=fs)

# Legend
plt.legend(frameon=False, loc="best", fontsize=fs)

# Save it
fig9.savefig("test9.pdf")
```
