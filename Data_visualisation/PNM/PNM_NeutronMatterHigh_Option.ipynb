## Additional Data

### dan_pnm_asysoft.dat

This file contains Danielewicz et al. heavy ion collision (HIC) constraints for PNM. This data was shared by Jirina Stone. This file contains 2 columns: 

Total baryonic density $\rho$ [fm$^{-3}$] and pressure $P$ [MeVfm$^{-3}$].

```python
dan_pnm_soft = np.loadtxt(OTHERS_DATA+HIC+"dan_pnm_asysoft.dat")
print(dan_pnm_soft.shape)
```

### dan_pnm_asystiff.dat

Dido to the above, just stiff.

```python
dan_pnm_stiff = np.loadtxt(OTHERS_DATA+HIC+"dan_pnm_asystiff.dat")
print(dan_pnm_stiff.shape)
```

## NeutronMatterHIgh Option Data Files

### pnm_eos_high.dat
